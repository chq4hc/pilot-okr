# Risk Estimation

```python
import datetime
from enum import Enum

class OKRStatus(Enum):
    ON_TRACK = "On Track"
    BEHIND = "Behind"
    AT_RISK = "At Risk"
    POSTPONED = "Postponed"
    CLOSED = "Closed"

class OKRTracker:
    def __init__(self, name, start_date, end_date, initial_value=0, target_value=100):
        self.name = name
        self.start_date = start_date
        self.end_date = end_date
        self.initial_value = initial_value
        self.target_value = target_value
        self.actual_value = initial_value
        self.key_results = []
        
    def add_key_result(self, kr):
        self.key_results.append(kr)

    def calculate_expected_progress(self, current_date=None):
        """Viva Goals calculates expected % based on time elapsed."""
        if not current_date:
            current_date = datetime.date.today()
            
        total_duration = (self.end_date - self.start_date).days
        elapsed_duration = (current_date - self.start_date).days
        
        if elapsed_duration <= 0: return 0.0
        if elapsed_duration >= total_duration: return 100.0
        
        return (elapsed_duration / total_duration) * 100

    def get_actual_progress_percent(self):
        """Calculates actual % based on current value vs target."""
        total_delta = self.target_value - self.initial_value
        current_delta = self.actual_value - self.initial_value
        if total_delta == 0: return 0.0
        return min(100.0, max(0.0, (current_delta / total_delta) * 100))

    def derive_status(self):
        """
        Logic from documentation:
        - At Risk: (Expected - Actual) > 25%
        - Behind: (Expected - Actual) > 0% and <= 25%
        - On Track: (Actual >= Expected)
        """
        actual = self.get_actual_progress_percent()
        expected = self.calculate_expected_progress()
        diff = expected - actual

        if diff > 25:
            return OKRStatus.AT_RISK
        elif 0 < diff <= 25:
            return OKRStatus.BEHIND
        else:
            return OKRStatus.ON_TRACK

    def get_health_alerts(self):
        """Implements Viva Goals 'Needs Attention' and 'At Risk' alerts."""
        alerts = []
        actual = self.get_actual_progress_percent()
        expected = self.calculate_expected_progress()
        today = datetime.date.today()

        # Alert: Past due date
        if today > self.end_date and actual < 100:
            alerts.append("[At Risk] Objective is past its due date.")

        # Alert: Progress gap > 25%
        if (expected - actual) > 25:
            alerts.append("[At Risk] Progress is off by more than 25% from expected.")

        # Alert: High score (> 0.8 / 80%)
        if actual > 80:
            alerts.append("[Needs Attention] Score is higher than 0.8. Consider a more ambitious goal.")

        # Alert: No Key Results
        if not self.key_results:
            alerts.append("[Needs Attention] Objective isn't aligned: it doesn't have key results.")

        return alerts

```

```python
# --- Example Usage ---

# Define a quarter cycle
start = datetime.date(2023, 10, 1)
end = datetime.date(2023, 12, 31)

# Create an Objective
objective = OKRTracker("Win the Indy 500", start, end)

# Simulate a manual check-in (e.g., we are halfway through time, but only 20% done)
objective.actual_value = 20  # Out of 100

# Calculate Status
expected = objective.calculate_expected_progress()
actual = objective.get_actual_progress_percent()
status = objective.derive_status()

print(f"Objective: {objective.name}")
print(f"Expected Progress: {expected:.1f}%")
print(f"Actual Progress: {actual:.1f}%")
print(f"Status: {status.value}")

# Check for Health Indicators
for alert in objective.get_health_alerts():
    print(f"Alert: {alert}")
```
